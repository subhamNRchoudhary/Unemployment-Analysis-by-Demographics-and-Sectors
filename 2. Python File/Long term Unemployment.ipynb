```python
import pandas as pd

# Load the dataset
file_path = r"C:\Users\sunny\OneDrive\Desktop\All the projects\17. Long-Term-Unemployment-Statistics\Long-Term-Unemployment-Statistics.xlsx"
df = pd.read_excel(file_path)

# Display the first few rows of the dataframe to understand its structure
df.head()
```

# Line Plot

```python
import matplotlib.pyplot as plt

# Line Plot
plt.figure(figsize=(10, 6))
plt.plot(df['Age'], df['Unemployed'], marker='o', linestyle='--', color='r')
plt.title('Unemployment by Age')
plt.xlabel('Age')
plt.ylabel('Number of Unemployed')
plt.grid(True)
plt.show()
```

# Bar Plot

```python
# Bar Plot
plt.figure(figsize=(10, 6))
plt.bar(df['Age'], df['Unemployed'], color='skyblue')
plt.title('Unemployment by Age')
plt.xlabel('Age')
plt.ylabel('Number of Unemployed')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Unemployed'], bins=5, color='orange')
plt.title('Histogram of Unemployment')
plt.xlabel('Number of Unemployed')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Participation Rate (%)'], df['Unemployed'], color='green')
plt.title('Unemployment vs Participation Rate')
plt.xlabel('Participation Rate (%)')
plt.ylabel('Number of Unemployed')
plt.show()
```

```python
# Pie Chart
sizes = df.groupby('Region')['Unemployed'].sum()
labels = sizes.index  # This ensures labels and sizes have the same length

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Unemployment Distribution by Region')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(10, 6))
df.boxplot(column='Unemployed', by='Gender')
plt.title('Box Plot of Unemployment by Gender')
plt.suptitle('')
plt.xlabel('Gender')
plt.ylabel('Number of Unemployed')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = r"C:\Users\sunny\OneDrive\Desktop\All the projects\17. Long-Term-Unemployment-Statistics\Long-Term-Unemployment-Statistics.xlsx"
df = pd.read_excel(file_path)

# Strip any leading/trailing spaces from column names
df.columns = df.columns.str.strip()

# Aggregate the data to remove duplicates
aggregated_df = df.groupby(['Age', 'Region'])['Unemployed'].sum().reset_index()

# Create a pivot table for the heatmap
pivot_table = aggregated_df.pivot('Age', 'Region', 'Unemployed')

# Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, cmap='coolwarm')
plt.title('Heatmap of Unemployment')
plt.show()
```

```python
# Violin Plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Gender', y='Unemployed', data=df)
plt.title('Violin Plot of Unemployment by Gender')
plt.show()
```

```python
# Area Plot
plt.figure(figsize=(10, 6))
plt.fill_between(df['Age'], df['Unemployed'], color="skyblue", alpha=0.4)
plt.plot(df['Age'], df['Unemployed'], color="Slateblue", alpha=0.6)
plt.title('Area Plot of Unemployment by Age')
plt.xlabel('Age')
plt.ylabel('Number of Unemployed')
plt.show()
```

```python
# Error Bar Plot
plt.figure(figsize=(10, 6))
plt.errorbar(df['Age'], df['Unemployed'], yerr=10000, fmt='o')
plt.title('Error Bar Plot of Unemployment by Age')
plt.xlabel('Age')
plt.ylabel('Number of Unemployed')
plt.show()
```
